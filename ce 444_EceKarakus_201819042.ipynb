```python
j=data[data["all parameters"]>=1]
a=data[data["all parameters"]!=0]
dp=pd.concat([pd.merge(j,a)])
dp
```

```python
dp=dp.iloc[dp['STREET_NAME'].argsort()]
dp
```

```python
dm=dp[dp.duplicated('STREET_NAME')| dp.duplicated('STREET_NAME', keep='last')]
dm
```

```python
dt=dm.groupby(['STREET_NAME'])['all parameters'].agg('max')
dt
```

```python
dm = dm.merge(dt, on='STREET_NAME',suffixes=('', '_max'))
dm
```

```python
dm = dm[(dm['all parameters'])==(dm['all parameters_max'])].drop('all parameters_max', axis=1)
dm
```

```python
a=dm[dm["all parameters"]>6]
n=dm[dm["all parameters"]!=0]
du=pd.concat([pd.merge(a,n)])
du
```

```python
x=du["all parameters"]
y=du["STREET_NAME"] 
fig = plt.figure()
plt.scatter(x,y)
fig.suptitle('Selected Locations for Traffic Lights and Signs')
plt.xlabel('All Parameters')
plt.ylabel('Locations')
plt.legend('Locations')
```

```python
#the graph came out very crowded so I decided to select the locations which has more than 10 parameters at the same time. In other words the top 10 locations which has the most need for traffic lights and signs.
y=dm[dm["all parameters"]>=10]
u=dm[dm["all parameters"]!=0]
dn=pd.concat([pd.merge(y,u)])
dn
```

```python
x=dn["all parameters"]
y=dn["STREET_NAME"] 
fig = plt.figure()
plt.scatter(x,y)
fig.suptitle('Top 10 Locations Which Has the Need for Traffic Lights and Signs')
plt.xlabel('All Parameters')
plt.ylabel('Locations')
plt.legend('Locations')
```
